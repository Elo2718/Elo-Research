# Rough Guide a v1 model.
0. Make a model to predict PTS 
0. For v1, dont use `2P, 3P, FG, or FT`.  Instead use `2PA, 3PA, FTA`.  
0. Include any other cols use see fit.  


````python
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler

target = 'PTS'
exclude_features = ['Team', 'Playoffs', 'W', 'PTS', 'oppPTS', '2P', '3P', 'FG']
features = [col for col in df.columns if col not in exclude_features]

# Function to calculate adjusted R
def adjusted_r2(r2, n, k):
    return 1 - (1 - r2) * (n - 1) / (n - k - 1)

selected_features = []
current_r2_adj = 0

# Feature selection loop (Forward Selection)
while True:
    improvements = []  
    for feature in features:
        if feature not in selected_features:
            temp_features = selected_features + [feature]
            X_temp = df[temp_features]
            y = df[target]
            
            scaler = StandardScaler()
            X_temp_normalized = scaler.fit_transform(X_temp)
            
            model = LinearRegression().fit(X_temp_normalized, y)
            y_pred = model.predict(X_temp_normalized)
            r2 = r2_score(y, y_pred)
            r2_adj = adjusted_r2(r2, len(y), len(temp_features))
            
            improvements.append((feature, r2_adj))
    
    best_feature, best_r2_adj = max(improvements, key=lambda x: x[1])
    
    if best_r2_adj > current_r2_adj:
        selected_features.append(best_feature)
        current_r2_adj = best_r2_adj
        print(f"Added {best_feature}, Adjusted R: {current_r2_adj:.4f}")
    else:
        print("No further improvement in Adjusted R.")
        break

print("\nFinal selected features:", selected_features)

if selected_features:
    correlation_matrix = df[selected_features].corr()
    fig, ax = plt.subplots(figsize=(10, 8))
    cax = ax.matshow(correlation_matrix, cmap='coolwarm')
    fig.colorbar(cax)
    ax.set_xticks(np.arange(len(selected_features)))
    ax.set_yticks(np.arange(len(selected_features)))
    ax.set_xticklabels(selected_features, rotation=45, ha='left')
    ax.set_yticklabels(selected_features)
    for i in range(len(selected_features)):
        for j in range(len(selected_features)):
            plt.text(j, i, f"{correlation_matrix.iloc[i, j]:.2f}", ha="center", va="center", color="black")
    plt.title("Correlation Matrix of Selected Features")
    plt.show()

target = 'PTS'

X = df[selected_features]
y = df[target]

scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)

model = LinearRegression().fit(X_normalized, y)

intercept = model.intercept_
coefficients = model.coef_

formula = f"PTS = {intercept:.4f}"
for feature, coef in zip(selected_features, coefficients):
    formula += f" + ({coef:.4f} * {feature})"

print("Regression Formula:")
print(formula)
````
